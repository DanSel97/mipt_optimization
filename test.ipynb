```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook

def f_1(x):
    return 3/(x-1)

def f_2(x):
    return 3/(x+2) - 4

def test(x):
    k = -4/3
    b = -8/3
    return k*x + b
x = np.linspace(-0.99, 0)
plt.plot(x, f_1(x))
plt.plot(x, f_2(x))
plt.plot(x, test(x))
```
